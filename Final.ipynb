````python
def clean_and_convert(df):
    if 'Date' in df.columns:
        df['Date'] = pd.to_datetime(df['Date'], errors= 'coerce')

    if 'Time' in df.columns:
        df['Hour'] = df['Time'].astype(str).str.split('.').str[0].astype(float)

    for col in df.select_dtypes(include=['object']).columns:
        df[col] = df[col].str.replace(',', '.', regex= True).astype(float, errors= 'ignore')

    return df
    pass  # TODO

# اعمال متد روی دیتاست
df_copy = clean_and_convert(df_copy)
````

````python
# TODO: کتابخانه های مورد نیاز را ایمپورت کنید
from sklearn.feature_selection import f_classif
from sklearn.feature_selection import mutual_info_regression
from sklearn.decomposition import PCA
````

````python
# TODO : متدی بتویسید که همبستگی چند جمله ایی را بررسی و مدیریت کند (Multicolinearity)
def multicorr(df_copy, target, threshold=0.8, anova_threshold=0.05):
    corr_matrix = df_copy.corr().abs()
    upper_triangle = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    drop_corr = [column for column in upper_triangle.columns if any(upper_triangle[column] > threshold)]
    X = df_copy.drop(columns=target)
    y = df_copy[target]
    F_values, p_values = f_classif(X, y)
    drop_anova = X.columns[p_values > threshold]
    drop = list(set(drop_corr + list(drop_anova)))
    df_reduced = df_copy.drop(columns=drop)
    return df_reduced, drop_corr, drop_anova
df_reduced, dropped_corr, dropped_anova = multicorr(df_copy, target='CO(GT)', threshold = 0.8, anova_threshold=0.05)
df_reduced
````

````python
# TODO : ویژگی هایی را به کمک روش Mutual Info انتخاب کنید.
def mutual_info(df_copy, target, threshold = 0.01):
    X = df_copy.drop(columns=target)
    y = df_copy[target]
    mi = mutual_info_regression(X, y)
    mi_series = pd.Series(mi, index=X.columns)
    selected_features = mi_series[mi_series > threshold].index.tolist()
    return df_copy[selected_features + [target]]
df_selected = mutual_info(df_reduced, target='CO(GT)', threshold=0.01)
df_selected
````

````python
# TODO : کتابخانه های مورد نیاز را ایمپورت کنید
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import learning_curve
````

````python
# TODO: چهار مدل اولیه را آموزش دهید و آنها را ارزیابی کنید
X = df_copy.drop(columns=['CO(GT)'])
y = df_copy['CO(GT)']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

models = {
          'Gradient Boosting': GradientBoostingRegressor(random_state=42),
          'XGBoost': XGBRegressor(random_state=42),
          'LightGBM': LGBMRegressor(random_state=42),
          'SVR': SVR()}
results = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    results[name] = {'Model': model, 'y_pred': y_pred}

for name, res in results.items():
    mse=mean_squared_error(y_test, res['y_pred'])
    mae=mean_absolute_error(y_test, res['y_pred'])
    r2 = r2_score(y_test, res['y_pred'])
    print(f"{name}:\n Mse: {mse:.4f}, MAE: {mae:.4f}, R2: {r2:.4f}\n")
````

````python
# TODO : نمودار Learning Curve هر مدل را رسم کنید
plt.figure(figsize=(14, 10))
for i, (name, res) in enumerate(results.items()):
    plt.subplot(2, 2, i+1)
    train_sizes, train_scores, val_scores = learning_curve(res['Model'], X, y, cv=5, scoring='neg_mean_squared_error')
    train_mean = -train_scores.mean(axis=1)
    val_mean = -val_scores.mean(axis=1)
    plt.plot(train_sizes, train_mean, label='Train')
    plt.plot(train_sizes, val_mean, label='Validation')
    plt.title(name)
    plt.xlabel('Training Size')
    plt.ylabel('MSE')
    plt.legend()
plt.tight_layout()
plt.show()
````
